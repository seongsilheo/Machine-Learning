# 3.1.3 3차 함수 f(x) = (x - 2)x(x + 2) 그리기

```python
# 리스트 2-(1)
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
def f(x):
    return (x-2)*x*(x+2)
```

```python
# 리스트 2-(2)
print(f(1))
```

x는 ndarry 배열이며 각각에 대응한 f를 한꺼번에 nearry로 돌려준다. 벡터의 사칙 연산은 각 요소마다 이루어지는 성질이 있기 때문이다.

```python
# 리스트 2-(3)
print(f(np.array([1,2,3])))
```

## 3.1.6 그래프를 작성하기
그런데 이 그래프는 정말로 x = -2, 0, 2일 때 f(x)가 0의 값을 값는지 여부를 알기 어렵다. 또한, 함수의 계수가 변경될 때 그래프가 어떻게 변하는지도 궁금한 분도 있을 것이다. 조금 손질한 그래프를 그려보자.

```python
#리스트 2-(7)

def f2(x, w):
    return (x-w)*x*(x+2) #함수 정의

x = np.linspace(-3,3,100) #x를 100 분할하기

#차트 묘사
plt.plot(x, f2(x,2), color = 'black', label = '$w=2$')
plt.plot(x, f2(x,1), color = 'cornflowerblue', label = '$w=1$')
plt.legend(loc="upper left") #범례 표시
plt.ylim(-15,15) #y축의 범위
plt.title('$f_2(x)$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.grid(True)
plt.show()
```

# 3.2 3차원 그래프 그리기
## 3.2.1 이변수 함수
두 변수의 함수, 즉 이변수 함수를 그림으로 나타내면 어떻게 하면 좋을까? 예를 들어, 
f(x0,x1) = (2x0^2 + x1^2)exp(-(2x0^2 + x1^2)
다음과 같은 함수의 그래프 형태를 알아보려면 어떻게 하면 좋을까? 
먼저, 앞의 함수를 **f3**으로 정의한다. 그리고 다양한 **x0**과 **x1**값에 대한 **f3**의 값을 계산한다.

```python
# 리스트 3-(1)
import numpy as np
import matplotlib.pyplot as plt

# 함수 f3을 정의
def f3(x0,x1):
    r = 2 * x0**2 + x1**2
    ans = r * np.exp(-r)
    return ans

# x0, x1에서 각각 f3을 계산
xn = 9
x0 = np.linspace(-2, 2, xn)
x1 = np.linspace(-2, 2, xn)
y = np.zeros((len(x0), len(x1)))  # x0길이 * x1길이의 행렬을 만들되 모든 값을 0로 초기화시킴
for i0 in range(xn):
    for i1 in range(xn):
        y[i1,i0] = f3(x0[i0], x1[i1])
```

```python
# 리스트 3-(2)
print(x0)
print(x1)
```

```python
# 리스트 3-(3)
print(np.round(y,1))
```

중심과 주변이 0으로, 도너츠 같은 모습을 하고 있다.

## 3.2.3 함수의 표면을 표시: surface

```python
#리스트 3-(5)
from mpl_toolkits.mplot3d import Axes3D #(A)

xx0, xx1 = np.meshgrid(x0,x1) #(B)
ax = plt.subplot(1,1,1,projection='3d') #(C)
ax.plot_surface(xx0, xx1, y, rstride=1, cstride=1, alpha=0.3, color='blue', edgecolor='black' ) #(D)
ax.set_zticks((0,0.2)) #(E)
ax.view_init(75,-95) #(F)
plt.show()
```

 surface를 표시하는 부분은 (D)에서 **ax.plot_surface(xx0, xx1, y)**이다. 옵션인 **rstride**와 **cstride**에 자연수를 부여해 가로 및 세로로 몇 개의 선을 긋는지 지정할 수 있다. 수가 적을수록 선의 간격이 조밀해진다.**alpha**는 면의 투명도를 지정하는 옵션이다. 0에서 1사이의 실수로 면의 투명도를 지정하며, 1에 가까울수록 불투명해진다. <br/><br/>
 z축의 눈금은 기본 상태로 표시하면 숫자가 겹쳐버린다. 그래서 **ax.set_zticks((0,0.2))** 명령으로 z의 눈금을 0과 0.2로 제한하고 있다. <br/><br/>
  (F)의 **ax.view_init(인수1, 인수2)**sms 3차원 그래프의 방향을 조절한다. **'인수1'**은 상화 회전 각도를 나타낸다. 0을 지정하면 옆에서 본 그래프가 되고, 90을 지정하면 위에서 본 그래프가 된다. **'인수2'**는 좌우 회전 각도를 나타내며 양수를 입력하면 시계방향으로, 음수를 부여하면 시계 반대 방향으로 차트가 회전한다.

## 3.2.4 등고선으로 표시: Contour
함수의 높이를 알아보려면 등고선 plot(일반적으로 둘 이상 변수 사이의 관계를 나타내는 그래프로, 데이터셋을 나타내는 그래픽 기술)이 편리하다.

```python
# 리스트 3-(9)
xn = 50
x0 = np.linspace(-2,2,xn)
x1 = np.linspace(-2,2,xn)

y = np.zeros((len(x0), len(x1)))
for i0 in range(xn):
    for i1 in range(xn):
        y[i1, i0] = f3(x0[i0], x1[i1])
        
xx0, xx1 = np.meshgrid(x0,x1) # (A)

plt.figure(1, figsize=(4,4))
cont = plt.contour(xx0, xx1, y, 5, colors='black') # (B)
cont.clabel(fmt='%3.2f', fontsize = 8)
plt.show()
```

(B)의 **plt.contour(xx0, xx1, y, 5, colors='black')**로 등고선 플롯을 작성한다. 5는 표시하는 높이를 5단계로 지시한다. <br/><br/>
**plt.contour**의 반환값을 **cont**에 저장하고, **cont.clabel(fmt="%3.2f',fontsize=8)**로 설정하여,각 등고선에 숫자를 넣을 수 있다(C).**fmt='%3.2f'**로 숫자 형식을 지정하고, **fontsize**로 문자의 크기를 결정한다.
